```python
### boucle for
list_weight= [0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1]
accuracy=[]

for weight in list_weight:

    full_test["angry_mean"]=(weight*full_test["angry"]+(1-weight)*full_test["angry_raph"])

    full_test["disgust_mean"]=(weight*full_test["disgust"]+(1-weight)*full_test["disgust_raph"])

    full_test["fear_mean"]=(weight*full_test["fear"]+(1-weight)*full_test["fear_raph"])

    full_test["happy_mean"]=(weight*full_test["happy"]+(1-weight)*full_test["happy_raph"])

    full_test["neutral_mean"]=(weight*full_test["neutral"]+(1-weight)*full_test["neutral_raph"])

    full_test["sad_mean"]=(weight*full_test["sad"]+(1-weight)*full_test["sad_raph"])
    
    full_test["emotion_predict"]=full_test[['angry_mean','disgust_mean','fear_mean',"happy_mean",'neutral_mean','sad_mean']].idxmax(axis=1)
    full_test["emotion_true"]=full_test["emotion"].map(classes)
    
    accuracy.append(accuracy_score(full_test["emotion_true"], full_test["emotion_predict"]))
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
from PIL import Image
import io

def get_array(audio_file):
    data, sample_rate = librosa.load(audio_file, sr=44100)
    fig = plt.figure
    sgram = librosa.stft(data)
    sgram_mag, _ = librosa.magphase(sgram)
    sample_rate = 44100
    mel_scale_sgram = librosa.feature.melspectrogram(S=sgram_mag, sr=sample_rate)
    mel_sgram = librosa.amplitude_to_db(mel_scale_sgram, ref=np.min)
    librosa.display.specshow(mel_sgram, sr=sample_rate, x_axis='time', y_axis='mel')
    plt.axis('off') 
    buf = io.BytesIO()
    plt.savefig(buf, format='png')
    buf.seek(0)
    image = Image.open(buf).convert('RGB') 
    resized_image = image.resize((256,256))
    X_sample = np.array(resized_image)
    X_sample = np.expand_dims(X_sample, axis = 0)
    return X_sample 
```

```python
def convert_mp3(path_myrecording):
    # convert mp3 to wav file
    subprocess.call(['ffmpeg', '-i', path_myrecording,
             '../samples/converted_to_wav_file.wav'])
    
    X_1 = sound_to_number("../samples/converted_to_wav_file.wav")
    
    X_2 = get_array("../samples/converted_to_wav_file.wav")
    
    os.remove("../samples/converted_to_wav_file.wav")
    
    return X_1, X_2
```

```python
x1, x2= convert_mp3("../recordings.mp3")
```

```python
from tensorflow.keras.models import load_model
import joblib
from speech_emotion_reco.raph_data import get_array
from speech_emotion_reco.mateo_preprocess import sound_to_number
import pandas as pd

def combine_predict(X_1, X_2):
    X_1 = pd.DataFrame(X_1)

    model_1= joblib.load("../models/mlp_model.joblib")
    model_2= load_model("../models/CNN_model.hdf5")
    proba1= pd.DataFrame(model_1.predict_proba(X_1.T))
    proba2= pd.DataFrame(model_2.predict(X_2))

    proba1.rename(columns={0: "angry", 1: "disgust",2:"fear",3:"happy",4:"neutral",5:"sad"}, inplace=True)
    proba2.rename(columns={0: "happy", 1: "sad",2:"fear",3:"disgust",4:"angry",5:"neutral"}, inplace=True)
    
    combined_proba= (proba1+proba2)/2
    combined_proba.sort_values(by=0,ascending=False, axis=1, inplace=True)
    proba_first_three= combined_proba.iloc[:,0:3]
    proba_dict={"emotion1":[proba_first_three.columns[0],proba_first_three.iloc[0,0]],"emotion2":[proba_first_three.columns[1],proba_first_three.iloc[0,1]],"emotion3":[proba_first_three.columns[2],proba_first_three.iloc[0,2]]}

    #combined_proba = list(combined_proba).sort(reverse=True)
    #return combined_proba[0:3]
    #emotion_predict=combined_proba[['angry','disgust','fear',"happy",'neutral','sad']].idxmax(axis=1)
    #return {"prediction":emotion_predict}
    return proba_dict
```
