```python
import pandas as pd
import numpy as np
```

```python
test= pd.read_csv("../speech_emotion_reco/data/df_test.csv")
```

```python
combined_df=pd.read_csv("../speech_emotion_reco/data/test_combined_proba.csv")
```

```python
combined_df.head()
```

```python
combined_df.shape
```

```python
full_test= combined_df.merge(test, on="observation")
```

```python
full_test.head()
```

```python
full_test["angry_mean"]=(full_test["angry"]+full_test["angry_raph"])/2
```

```python
full_test["disgust_mean"]=(full_test["disgust"]+full_test["disgust_raph"])/2
```

```python
full_test["fear_mean"]=(full_test["fear"]+full_test["fear_raph"])/2
```

```python
full_test["happy_mean"]=(full_test["happy"]+full_test["happy_raph"])/2
```

```python
full_test["neutral_mean"]=(full_test["neutral"]+full_test["neutral_raph"])/2
```

```python
full_test["sad_mean"]=(full_test["sad"]+full_test["sad_raph"])/2
```

```python
full_test.head()
```

```python
full_test["emotion_predict"]=full_test[['angry_mean','disgust_mean','fear_mean',"happy_mean",'neutral_mean','sad_mean']].idxmax(axis=1)
```

```python
full_test.head()
```

```python
classes = {0:'happy_mean',
          1:'sad_mean',
          2:'fear_mean',
          3:'disgust_mean',
          4:'angry_mean',
          5:'neutral_mean'}
```

```python
full_test["emotion_true"]=full_test["emotion"].map(classes)
```

```python
full_test.head()
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(full_test["emotion_true"], full_test["emotion_predict"])
```

```python
### Proba que Mateo
full_test["emotion_predict_mateo"]=full_test[['angry','disgust','fear',"happy",'neutral','sad']].idxmax(axis=1)
```

```python
accuracy_score(full_test["emotion_true"], full_test["emotion_predict_mateo"])
```

```python
full_test["emotion_predict_mateo"]=full_test["emotion_predict_mateo"]+"_mean"
```

```python
full_test["emotion_predict_raph"]=full_test[['angry_raph','disgust_raph','fear_raph',"happy_raph",'neutral_raph','sad_raph']].idxmax(axis=1)
```

```python
full_test.head()
```

```python
class_raph={"neutral_raph":"neutral_mean",
           "sad_raph":"sad_mean",
           "happy_raph":"happy_mean",
           "disgust_raph":"disgust_mean",
           "fear_raph":"fear_mean",
           "angry_raph":"angry_mean"}
```

```python
full_test["emotion_predict_raph"]=full_test["emotion_predict_raph"].map(class_raph)
```

```python
accuracy_score(full_test["emotion_true"], full_test["emotion_predict_raph"])
```

```python
### boucle for
list_weight= [0.9,0.8,0.7,0.6,0.5,0.4,0.3,0.2,0.1]
accuracy=[]

for weight in list_weight:

    full_test["angry_mean"]=(weight*full_test["angry"]+(1-weight)*full_test["angry_raph"])

    full_test["disgust_mean"]=(weight*full_test["disgust"]+(1-weight)*full_test["disgust_raph"])

    full_test["fear_mean"]=(weight*full_test["fear"]+(1-weight)*full_test["fear_raph"])

    full_test["happy_mean"]=(weight*full_test["happy"]+(1-weight)*full_test["happy_raph"])

    full_test["neutral_mean"]=(weight*full_test["neutral"]+(1-weight)*full_test["neutral_raph"])

    full_test["sad_mean"]=(weight*full_test["sad"]+(1-weight)*full_test["sad_raph"])
    
    full_test["emotion_predict"]=full_test[['angry_mean','disgust_mean','fear_mean',"happy_mean",'neutral_mean','sad_mean']].idxmax(axis=1)
    full_test["emotion_true"]=full_test["emotion"].map(classes)
    
    accuracy.append(accuracy_score(full_test["emotion_true"], full_test["emotion_predict"]))
```
